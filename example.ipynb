```python
import google.generativeai as genai
import re

# Configure with your Google AI Studio API key
API_KEY = "AIzaSyCO3D7McHLJ-cx3UFLs7af3zsZEFxqp3hg"  # Replace with your actual API key
genai.configure(api_key=API_KEY)

def analyze_lead_potential(interests):
    """
    Analyzes customer interests and returns likelihood percentage
    """
    model = genai.GenerativeModel('gemini-2.5-flash')
    
    prompt = f"""
    Act as an expert lead scoring analyst. Evaluate how well these customer interests align with our business:
    {interests}

    Our company specializes in: 
    - Software development services
    - Cloud computing solutions
    - AI/ML implementation
    - Digital transformation consulting

    Output format: 
    "Based on interest alignment, this lead has a [X]% likelihood of becoming a client. 
    Key factors: [concise bullet points]"
    """
    
    response = model.generate_content(prompt)
    return response.text

def extract_percentage(text):
    """Extracts percentage value from response text"""
    match = re.search(r'(\d+)%', text)
    return int(match.group(1)) if match else None

if __name__ == "__main__":
    # Get user input
    interests = input("Enter customer interests: ")
    
    # Generate analysis
    result = analyze_lead_potential(interests)
    
    # Extract and display percentage
    likelihood = extract_percentage(result)
    
    if likelihood is not None:
        print(f"\nLead Conversion Probability: {likelihood}%")
        print(f"Detailed Analysis:\n{result}")
    else:
        print("Error: Could not determine likelihood percentage")
        print(f"Full response: {result}")
```
